# iPhone sales analysis in India

```python
highest_rated=data.sort_values(by = ["Star Rating"], ascending = False)
highest_rated = highest_rated.head(10)
```

```python
highest_rated["Product Name"]
```

# Number of ratings of the highest rated iPhone on Flipkart

```python
iphones= highest_rated["Product Name"].value_counts()
x= iphones.index
y= highest_rated["Number Of Ratings"]
plt.barh(x,y)
plt.title("Number of reviews of the highest rated iPhones")
plt.grid(True)
plt.show()
```

```python
iphones
```

# Number of reviews of the highest rated iPhones

```python
iphones= highest_rated["Product Name"].value_counts()
x= iphones.index
y= highest_rated["Number Of Reviews"]
plt.bar(x,y,color="r")
plt.title("Number of reviews of the highest rated iPhones")
plt.xticks(x,rotation ='vertical') 
plt.show()
```

# Relationship B/W Sale Price and Number of Ratings

```python
x= data["Number Of Ratings"]
y=data["Sale Price"]
plt.scatter(x,y)
plt.title("Relationship B/W Sale Price and Number of Ratings")
plt.show()
```

# Number of Reviews of Highest Rated iPhones

```python
x= highest_rated["Product Name"]
y=highest_rated["Number Of Reviews"]
plt.bar(x,y)
plt.title("Number of Reviews of Highest Rated iPhones")
plt.xticks(x,rotation ='vertical') 
plt.show()
```

# Relationship b/w Discount Percentage and Number of Rating

```python
x= data["Number Of Ratings"]
y=data["Discount Percentage"]

plt.scatter(x,y)
plt.title("Relationship b/w Discount Percentage and Number of Rating")
plt.show()
```

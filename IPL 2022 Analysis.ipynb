# Project 2- IPL 2022 Analysis


```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("matches.csv")
```

```python
df
```

# Number of matches won by each team in ipl 2022

```python
# Calculate the number of matches won by each team
match_winners = df['match_winner'].value_counts()
plt.bar(match_winners.index, match_winners.values)
plt.title("Number of Matches Won in IPL 2022")
plt.xlabel("Teams")
plt.ylabel("Number of Wins")
plt.xticks(rotation=45)
plt.show()
```

# Best Bowling 

```python
best_bowling = df['best_bowling'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(best_bowling.index, best_bowling.values)
plt.title("Number of Matches Won in IPL 2022")
plt.xticks(rotation=90)
plt.show()
```

# Player of match

```python

potm = df['player_of_the_match'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(potm.index, potm.values, color = "gold")
plt.title("Player of the match in IPL 2022")
plt.xticks(rotation=90)
plt.show()
```

# Top Scorer

```python
x=df["top_scorer"]
y=df["highscore"]
plt.figure(figsize=(10, 6))
plt.bar(x,y,color="orange")
plt.title("Topscorer in every match of IPL 2022")
plt.xticks(rotation=90)
plt.show()
```

```python
df['won_by'] = df['won_by'].map({'Wickets': "Chasing", 'Runs': 'Defending'})
won_by = df["won_by"].value_counts()
labels = won_by.index
counts = won_by.values
colors = ['red', 'green']

plt.pie(counts, labels=labels, colors=colors, autopct='%1.1f%%',
        startangle=140, textprops={'fontsize': 14}, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 3})
plt.title("Number of matches won by defending and chasing", fontsize=18)
plt.show()
```
